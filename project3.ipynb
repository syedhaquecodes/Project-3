```
#acquiring all the data needed for the graphs
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("sample_data/WomenEntrepreneurs.csv")
members = df[df["EuropeanUnionMembership"] == "Member"]
nonMembers = df[df["EuropeanUnionMembership"] == "Not Member"]
developCountries = df[df["levelofdev"] == "Developed"]
developingCountries = df[df["levelofdev"] == "Developing"]
avgIndexDevelop = developCountries["WomenEntrepreneurshipIndex"].mean()
avgIndexDeveloping = developingCountries["WomenEntrepreneurshipIndex"].mean()
```

```
#members female participation rate
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
memberCountries = members["Country"] 
memberFemaleWorkForce = members["FemaleLaborForceParticipationRate"]
ax.bar(memberCountries,memberFemaleWorkForce)
plt.ylabel("Female Labor Participation Rate")
plt.title("European Union Members Female Participation Rate")
plt.xticks(rotation=45)
plt.ylim(5,85 )
plt.show()
```

```
#nonmembers female participation rate
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
nonMemberCountries = nonMembers["Country"] 
nonMemberFemaleWorkForce = nonMembers["FemaleLaborForceParticipationRate"]
ax.bar(nonMemberCountries,nonMemberFemaleWorkForce)
plt.ylabel("Female Labor Participation Rate")
plt.title("NON European Union Members Female Participation Rate")
plt.xticks(rotation=90)
plt.ylim(5,85 )
plt.show()
```

```
countries = developCountries["Country"]
countriesIndexes = developCountries["WomenEntrepreneurshipIndex"]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(countries, countriesIndexes, color='r')
plt.axhline(y=avgIndexDevelop, color='b', linestyle='-')
plt.xticks(rotation=75)
ax.set_xlabel('Developed Country')
ax.set_ylabel('WomenEntrepreneurshipIndex')
ax.set_title('Women Entrepreneurship Index Developed in Developed Countries')
plt.show()
```

```
devCountries = developingCountries["Country"]
devCountriesIndexes = developingCountries["WomenEntrepreneurshipIndex"]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(devCountries, devCountriesIndexes, color='g')
plt.axhline(y=avgIndexDeveloping, color='r', linestyle='-')
plt.xticks(rotation=80)
ax.set_xlabel('Developing Country')
ax.set_ylabel('WomenEntrepreneurshipIndex')
ax.set_title('Women Entrepreneurship Index Developed in Developing Countries')
plt.show()
```
